```
# accuracy result for Decision Tree Classifier
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
iris = load_iris()
X = iris.data 
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
tree_clf = DecisionTreeClassifier(criterion='gini',max_depth=5) 
tree_clf.fit(X_train, y_train)
y_pred = tree_clf.predict(X_test) 
print('accuracy result for Decision Tree Classifier:',accuracy_score(y_test, y_pred, normalize=True).round(4))
```

```
# the generated tree form the decision tree classifier
from sklearn.tree import export_graphviz 
export_graphviz(
tree_clf,
out_file="iris_tree.dot", feature_names=iris.feature_names, class_names=iris.target_names, rounded=True,
filled=True
)
import graphviz
with open("iris_tree.dot") as f:
  dot_graph = f.read() 
graphviz.Source(dot_graph)
```

```
# accuracy result for KNN Classifier
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
iris = load_iris()
X = iris.data 
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
knn_clf = KNeighborsClassifier(15) 
knn_clf.fit(X_train, y_train)
y_pred = knn_clf.predict(X_test) 
print('accuracy result for KNN Classifier:', accuracy_score(y_test, y_pred, normalize=True))
```

```
# show how the accuracy changes when K ranges from 11 to 20
accuracy_list = []
for k in range(11,21):
	knn_clf = KNeighborsClassifier(k) 
	knn_clf.fit(X_train, y_train)
	y_pred = knn_clf.predict(X_test)
	accuracy =  accuracy_score(y_test, y_pred, normalize=True).round(3)
	accuracy_list.append(accuracy)

import matplotlib.pyplot as plt
def vizualizeData(K, accuracy_score):
	scatter = plt.scatter(K,accuracy_score) 
	plt.ylabel('accuracy_score')
	plt.xlabel('K')
	plt.title('how the accuracy changes using 10 different values for K',fontsize = 12, color = 'red')
	plt.show()
vizualizeData(list(range(11, 21)), accuracy_list)
```

```
# show how the accuracy changes when max_depth ranges from 3 to 12
accuracy_list = []
for k in range(3,13):
	tree_clf = DecisionTreeClassifier(criterion='gini',max_depth=k) 
	tree_clf.fit(X_train, y_train)
	y_pred = tree_clf.predict(X_test)
	accuracy =  accuracy_score(y_test, y_pred, normalize=True).round(3)
	accuracy_list.append(accuracy)

import matplotlib.pyplot as plt
def vizualizeData(max_depth, accuracy_score):
	scatter = plt.scatter(max_depth,accuracy_score) 
	plt.ylabel('accuracy_score')
	plt.xlabel('max_depth')
	plt.title('how the accuracy changes using 10 different values for max_depth',fontsize = 12, color = 'red')
	plt.show()
 

vizualizeData(list(range(3, 13)), accuracy_list)
```
